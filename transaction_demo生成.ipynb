```python
import time
import random

with open('transaction_data_v3.txt', 'w', encoding='utf-8') as f:
    for i in range(100):
        transaction_id=int(i+1)
        member_id=10000+random.randint(1, 20)
        


        a1=(2018,1,1,0,0,0,0,0,0)              #设置开始日期时间元组（1976-01-01 00：00：00）
        a2=(2018,5,31,23,59,59,0,0,0)    #设置结束日期时间元组（1990-12-31 23：59：59）

        start=time.mktime(a1)    #生成开始时间戳
        end=time.mktime(a2)      #生成结束时间戳

#随机生成10个日期字符串

        t=random.randint(start,end)    #在开始和结束时间戳中随机取出一个
        date_touple=time.localtime(t)          #将时间戳生成时间元组
        date=time.strftime("%Y-%m-%d",date_touple)  #将时间元组转成格式化字符串（1976-05-21）
        
        f.write(str(transaction_id)+','+str(member_id)+','+date+','+'123'+'\n')

with open('transaction_data_v3.txt', 'r', encoding='utf-8') as f:
    print(f.read())
        
```
